```python
import numpy as np
import matplotlib.pylab as plt
plt.rc('font',family='Times New Roman')
fig = plt.figure()
```

```python
# vorticity

# mid 
V1 = np.loadtxt('vorticity/midaxis/9.xy',skiprows=2)
V2 = np.loadtxt('vorticity/midaxis/13.xy',skiprows=2)
V3 = np.loadtxt('vorticity/midaxis/17.xy',skiprows=2)
V4 = np.loadtxt('vorticity/midaxis/21.xy',skiprows=2)
V5 = np.loadtxt('vorticity/midaxis/24.xy',skiprows=2)

d1 = (V1[:,0]-100)/10.8
d2 = (V2[:,0]-100)/10.8
d3 = (V3[:,0]-100)/10.8
d4 = (V4[:,0]-100)/10.8
d5 = (V5[:,0]-100)/10.8

plt.plot(d1,V1[:,1],linestyle='-',c='k',label='SVF1')
# plt.plot(d2,V2[:,1],linestyle='-',c='b',label='SVF2')
# plt.plot(d3,V3[:,1],linestyle='-',c='b',label='SVF3')
# plt.plot(d4,V4[:,1],linestyle='-',c='m',label='SVF4')
plt.plot(d5,V5[:,1],linestyle='-',c='r',label='SVF5')

# range
plt.xlim([-5,9.5])
plt.ylim([-1,1])

# style
plt.tick_params(top=True,right=True)
plt.rcParams['ytick.direction']='in'
plt.rcParams['xtick.direction']='in'
plt.tick_params(labelsize=16)
axes=plt.subplot()
axes.minorticks_on()
plt.tick_params(which='major',length=6)
plt.legend(fontsize=12)
plt.xlabel('Distance(x/D)',fontsize=16)
plt.ylabel('Vorticity (s$^{-1}$) ',fontsize=16)

# canopy line 
x = [-0.5,-0.5]
y = [-2,3]
plt.plot(x,y,linestyle = '--',c='g')
x = [0.5,0.5]
y = [-2,3]
plt.plot(x,y,linestyle = '--',c='g')
```
